```python
new_df['total_arae'] = new_df['sqft_living']+new_df['sqft_lot']+new_df['sqft_above']+new_df['sqft_basement']
```

```python
new_df = new_df.drop(['sqft_living','sqft_lot','sqft_above','sqft_basement'], axis=1)
```

```python
new_df = new_df.rename(columns={'total_arae': 'Total_Area',
                               'yr_built' : 'Year_of_built'})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score ,mean_absolute_error, mean_squared_error,root_mean_squared_error
```

```python
X = new_df.iloc[:, 1:]
y = new_df.iloc[:,0:1]
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
print("r2_score - > ", r2_score(y_test,y_pred))
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test,y_pred))
```
